# Boston Home Values 2018 - 2023

```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 2018

```python
#Read in 2018
clean_data =pd.read_csv("Boston Property Assesment2018.csv")
clean_data.head()
```

```python
#change column name for consistency 
cleandf = clean_data.rename(columns={'AV_TOTAL': 'TOTAL_VALUE'}) 
```

```python
#filter out those NOT owner occupied
owner_occ = cleandf[cleandf['OWN_OCC'] == 'Y']
owner_occ
```

```python
#get median of home values
median2018 = owner_occ['TOTAL_VALUE'].median()
median2018

print("The median home value in 2018 was $",median2018)
```

## Line of Change 2018-2023

```python
#set x and y
years =['2018','2019','2020','2021','2022','2023']
median_all = [median2018,median2019,median2020,median2021,median2022,median2023]
```

```python
#Plot home value changes from 2018-2023
plt.plot(years, median_all, marker="o", label='line with marker')
plt.xlabel('Year')
plt.ylabel('Home Value')
plt.title('Median Home Values by Year')
plt.legend()
plt.show()
```
