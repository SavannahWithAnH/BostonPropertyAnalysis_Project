```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
clean_data1 =pd.read_csv("../Project1/Boston Property Assesment Data.csv")
```

```python
clean_data1.head()
clean_data1.columns
```

```python
useful_boston1 = clean_data1[['OWNER','CITY','OWNER','OWN_OCC','TOTAL_VALUE','GROSS_TAX','YR_BUILT']]
useful_boston1
```

```python
unique_cities = useful_boston1['CITY'].unique()
unique_cities
```

```python
boston_list = ['EAST BOSTON', 'BOSTON', 'SOUTH BOSTON']
```

```python
only_boston = useful_boston1[useful_boston1['CITY'].isin(boston_list)]
only_boston
```

```python
owner_occ = only_boston[only_boston['OWN_OCC'] == 'Y']
owner_occ
```

```python
size_bins = [0, 249999, 499999, 749999, 999999, 1249999, 1499999, 1749999, 1999999, 2249999, 24999999, 1000000000]
group_names = ["Under 250k", "250k-500k", "500k-750k", "750k-1M", "1M-1.25M", "1.25M-1.5M", "1.5M-1.75M", "1.75M-2M", "2M-2.25M", "2.25M-2.5M", "Over 2.5M"]
```

```python
owner_occ["Price Bins"] =  pd.cut(owner_occ["TOTAL_VALUE"], size_bins, labels=group_names)
owner_occ
```

```python
binned_prices = owner_occ.groupby(["Price Bins"]).count()["TOTAL_VALUE"]
binned_prices
```

```python
x_axis = np.arange(len(group_names))
tick_locations = [value for value in x_axis]

plt.bar(x_axis, binned_prices)

## color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, group_names, rotation="vertical")
plt.xlabel("Total Value")
plt.ylabel("Number of Homes")
plt.title("Boston Home Values")
plt.show()
```

```python
plt.figure(figsize=(10,6))
fig,ax=plt.subplots()    
plt.scatter(owner_occ["TOTAL_VALUE"],owner_occ["YR_BUILT"])
##start,end=ax.get_xlim()
ax.xaxis.set_ticks(np.arange(-117000, 3100000, 250000))
plt.ticklabel_format(style="plain")
plt.xticks(rotation="vertical")
plt.xlabel('Total Value')
plt.ylabel('Year Built')
plt.show()
```
